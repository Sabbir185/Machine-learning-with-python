```python
import pandas as pd
df = pd.read_csv("weight-height.csv")
df.head()
```

```python
df.shape
```

# Data preprocessing

```python
from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()
```

```python
df.Gender = label.fit_transform(df.Gender)
```

```python
df.head()
```

### split data into x, y

```python
x = df.drop('Weight', axis=1)
```

```python
y = df['Weight']
```

```python
y.head()
```

```python
x.head()
```

### split data into train and testing set

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.30, random_state=1) 
```

```python
xtrain.shape
```

# KNN Regression

```python
from sklearn.neighbors import KNeighborsRegressor

KNN_reg = KNeighborsRegressor()
```

```python
KNN_reg.fit(xtrain, ytrain)
```

```python
pred = KNN_reg.predict(xtest)
```

```python
pred
```

```python
KNN_reg.score(xtest, ytest)
```

#### Mean Square Error (MSE) & Mean Absolute Error(MAE)

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error

mse = mean_squared_error(ytest, pred)
mse
```

```python
mae = mean_absolute_error(ytest, pred)
mae
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
linear_reg = LinearRegression()
```

```python
linear_reg.fit(xtrain, ytrain)
```

```python
pred2 = linear_reg.predict(xtest)
pred2
```

```python
linear_reg.score(xtest, ytest)
```

```python
mse_linear_reg = mean_squared_error(ytest, pred2)
mse_linear_reg
```

```python
# for this dataset linear regression is performing better than KNN regression
```

# Hyper-parameter Optimization Using RandomSearchCV 
## For KNN Regression

```python
from sklearn.model_selection import RandomizedSearchCV
import numpy as np
```

```python
n_neighbors = [int(x) for x in np.linspace(start=20, stop=300, num=20)]
weights = ['uniform', 'distance']
algorithm = ['auto', 'ball_tree', 'kd_tree', 'brute']
```

```python
grids = {'n_neighbors': n_neighbors,
         'weights': weights,
         'algorithm': algorithm
        }
```

```python
KNN_reg2 = KNeighborsRegressor()
```

```python
rmcv = RandomizedSearchCV(KNN_reg2, grids, n_iter=100, cv=5)
```

```python
rmcv.fit(xtrain, ytrain)
```

```python
rmcv.cv_results_
```

```python
results = pd.DataFrame(rmcv.cv_results_)
results
```

```python
rmcv.best_params_
```

```python
rmcv.best_estimator_
```

```python
rmcv.best_score_
```

```python
KNN_reg3 = rmcv.best_estimator_     # model assign with best parameters

KNN_reg3
```

```python
KNN_reg3.score(xtest, ytest)  # KNN best score after tuning
```
