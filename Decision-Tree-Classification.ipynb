```python
import pandas as pd
df = pd.read_csv("tree.csv")  #import DataFrame 
df
```

```python
df.info()
```

```python
y = df["Wear Jacket?"]
y.head()
```

```python
x = df.drop("Wear Jacket?", axis=1)
x.head()
```

# Preprocessing

```python
from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()
x['Outlook'] = label.fit_transform(df["Outlook"])
x['Temprature'] = label.fit_transform(df["Temprature"])
x['Routine'] = label.fit_transform(df["Routine"])
```

```python
x.head()
```

```python
x = x.drop('Days', axis=1)  # here days column unnecessary
```

```python
x.head()
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier()
```

```python
clf.fit(x, y)  # train with full data
```

```python
clf.predict([[1, 0, 0]])  # sunny-> 1, cold -> 0, indoor -> 0
```

```python
clf.predict([[1, 1, 0]])  # sunny-> 1, warm -> 1, indoor -> 0
```

```python
clf.predict([[1, 0, 1]])  # sunny-> 1, cold -> 0, outdoor -> 1
```

# Tree structure show

```python
from sklearn import tree
tree.plot_tree(clf)
```
